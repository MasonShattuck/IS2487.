```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Polar_bear'
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    for link in soup.find_all('a'):
        href = link.get('href')
        if href and href.startswith('http'):
            print(href)
else:
    print("Failed to retrieve the webpage")
print("Number of links found:", len(href))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'TeamName': ['Reese', 'Barnes', 'Jones', 'Javez', 'Smith', 'Heath', 'Jacks', 'Weeks', 'Mitch', 'Rankle'],
    'HighestScore': [112, 103, 127, 99, 130, 107, 89, 120, 122, 90],
    'LowestScore': [47, 87, 70, 51, 100, 101, 38, 97, 110, 55]}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 5))
sns.lineplot(data=df.set_index('TeamName'), markers=True)
plt.title('Highest and Lowest Scores by Team')
plt.xlabel('Team Name')
plt.ylabel('Score')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
